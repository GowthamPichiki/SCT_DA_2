```python
import pandas as pd


# Load the dataset using 'latin1' encoding to handle special characters smoothly
df = pd.read_csv('superstore.csv', encoding='latin1')
print(f"Initial Dataset Shape: {df.shape}")

# Replace dots with underscores in column names for standard coding practices
df.columns = df.columns.str.replace('.', '_', regex=False)

# Drop the redundant/corrupted column 'è®°å½•æ•°'
if 'è®°å½•æ•°' in df.columns:
    df = df.drop(columns=['è®°å½•æ•°'])
    print("✓ Dropped redundant special-character column.")

# The original 'Order_Date' and 'Ship_Date' contain '00:00.0'. 
# We reconstruct valid dates using the 'Year' and 'weeknum' columns.
df['Order_Date'] = pd.to_datetime(
    df['Year'].astype(str) + '-W' + df['weeknum'].astype(str) + '-1', 
    format='%G-W%V-%u', 
    errors='coerce'
)

# Estimate 'Ship_Date' by adding a standard 4-day shipping turnaround to the Order_Date
df['Ship_Date'] = df['Order_Date'] + pd.to_timedelta(4, unit='D')
print("✓ Restored and cleaned corrupted date columns.")

# Check for and drop exact row duplicates
initial_count = len(df)
df = df.drop_duplicates()
final_count = len(df)
print(f"✓ Dropped {initial_count - final_count} duplicate rows.")


print("\n--- Final Cleaned Data Types (Sample) ---")
print(df[['Order_Date', 'Ship_Date', 'Sales', 'Profit', 'Category']].dtypes)

output_filename = 'Cleaned_Superstore.csv'
df.to_csv(output_filename, index=False)
print(f"\n🎉 Success! Your cleaned dataset has been saved as: '{output_filename}'")
```
